```python
%matplotlib notebook
import numpy as np
```

```python
def f_non_convex(x, y):
    return (1 - x / 2.5 + x ** 5 + y ** 3 ) * np.exp(-x ** 2 - y ** 2)

def f_convex(x, y):
    return 3/2*x**2 + 1/2*y**2
```

```python
X = np.arange(-3,3,0.03)
Y = np.arange(-3,3,0.03)
X, Y = np.meshgrid(X,Y)
Z = f_convex(X,Y)
Z, X.shape, Y.shape, Z.shape
```

```python
import matplotlib.pyplot as plt
from matplotlib import cm

plt.rcParams['figure.figsize'] = [5, 5]
plt.rcParams['figure.dpi'] = 200

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
surf = ax.plot_surface(X, Y, Z, cmap=cm.rainbow, linewidth=0, antialiased=False)
fig.colorbar(surf, aspect=10)
plt.show()
```
